## Redes acíclicas

Um exemplo clássico de uma rede acíclica dirigida é uma rede de citações de artigos, como a que foi referida na aula anterior. Ao escrever um artigo, o autor só pode citar outro artigo se este já tiver sido escrito, o que significa que todas as arestas de uma rede de citações apontam para trás no tempo, de artigos posteriores para artigos anteriores.

Não é óbvio, mas não deixa de ser verdade e pode ser demonstrado, que se uma rede for acíclica poderá ser desenhada da forma apresentada no slide anterior, com todas as arestas a apontar para baixo.

```python
G=nx.gnp_random_graph(10,0.5,directed=True)
DAG = nx.DiGraph([(u,v) for (u,v) in G.edges() if u < v])
nx.is_directed_acyclic_graph(DAG)
```

```python
nx.draw(DAG,pos = nx.circular_layout(DAG), arrowsize=20, node_color='white', edgecolors='black', node_size=500, 
        with_labels =True)
```

## Projeção de redes Bipartidas

```python
from networkx.algorithms import bipartite
B = nx.Graph()
B.add_nodes_from([1, 2, 3, 4], bipartite=0)
B.add_nodes_from(["a", "b", "c"], bipartite=1)
B.add_edges_from([(1, "a"), (1, "b"), (2, "b"), (2, "c"), (3, "c"), (4, "a")])
nx.is_bipartite(B)
```

```python
bottom_nodes, top_nodes = bipartite.sets(B)
nx.draw(B, nx.bipartite_layout(B, top_nodes, align='horizontal'), arrowsize=20, node_color='white', edgecolors='black', node_size=500, 
        with_labels =True)
```

```python
from matplotlib import pyplot as plt
fig, axs = plt.subplots(ncols=2)
G1 = bipartite.projected_graph(B, bottom_nodes)
G2 = bipartite.projected_graph(B, top_nodes)
nx.draw(G1, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True, ax=axs[0])
nx.draw(G2, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True, ax=axs[1])
plt.show()
```

# Grau

O grau de um nó numa rede não direcionada é o número de arestas ligadas a ele

$$k_i = \sum{n}_{j=1}A_{ij} $$

O número total 2m das pontas das m arestas de uma rede é dado por:

$$2m = \sum{n}_{i=1}k_{i}=\sum_{ij}A_{ij}$$

O grau médio de uma rede é:

$$c = \frac{1}{n}\sum{n}_{i=1}k_i=\frac{2m}{n}$$

## Cálculo do grau

```python
G = nx.erdos_renyi_graph(n = 10, p = 0.5)
nx.draw(G, nx.kamada_kawai_layout(G), arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True)
list(G.degree(list(range(10))))
```

# Caminhos (paths)


Caminhos mais curtos
Um caminho mais curto numa rede, por vezes também designado por caminho geodésico, é o caminho mais curto entre um determinado par de nós, ou seja, o caminho que atravessa o menor número de arestas.

A distância mais curta ou distância geodésica entre dois nós, muitas vezes chamada apenas de “distância”, é o comprimento do caminho mais curto em termos do número de arestas.

Os caminhos mais curtos e as distâncias mais curtas desempenham um papel importante em vários fenómenos de rede.

Caminhos mais curtos
Por exemplo, o efeito mundo pequeno é, na verdade, uma afirmação sobre distâncias mais curtas – que são surpreendentemente pequenas mesmo nas maiores redes. As distâncias mais curtas são importantes em redes de comunicação como a Internet, onde afetam a rapidez com que é possível levar dados de um nó para outro, ou em redes de transporte, como as redes aéreas, onde determinam quantos troços serão necessários para uma viagem.

É possível que não exista um caminho mais curto entre dois nós se os nós não estiverem ligados por qualquer rota através da rede (ou seja, se estiverem em “componentes” diferentes (ver a seguir).

O diâmetro de uma rede é o comprimento do “maior dos caminhos mais curtos.”

## Caminhos mais curtos

```python
G = nx.Graph()
nx.add_path(G, [0, 1, 2])
nx.add_path(G, [0, 10, 2])
nx.add_path(G, [0, 5, 2])
nx.add_path(G, [5, 7, 2])
nx.draw(G, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True)
print([p for p in nx.all_shortest_paths(G, source=0, target=2)])
```
